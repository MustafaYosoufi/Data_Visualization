```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/train.csv')
df2 = pd.read_csv('/content/test.csv')
if 'SalePrice' not in df2.columns:
    df2['SalePrice'] =0
df = pd.concat([df1,df2], axis = 0)
df = df.set_index('Id')
```

```python
df_objects = df[df.select_dtypes(include=['object']).columns]
df = df.drop(df[df_objects.isna().sum()[df_objects.isna().sum()>1100].index],axis=1)
```

```python
df_objects = df_objects.drop(columns=df_objects.isna().sum()[df_objects.isna().sum() > 1100].index)

#df_objects = df_objects.drop(df_objects[df_objects.isna().sum()[df_objects.isna().sum() >1100].index)

df_objects = df_objects.fillna('null')
df_object_encoded = pd.get_dummies(df_objects)
```

```python
df_objects['MSZoning'].value_counts()
```

```python
df_object_encoded.head()
```

```python
for i in df_object_encoded.columns:
    if 'null' in i:
        df_object_encoded=df_object_encoded.drop(i,axis=1)
        print(i)
```

```python
new_df = pd.concat([df,df_object_encoded],axis=1)
```

```python
len(new_df.columns)
```

```python
len(df.columns),len(df_object_encoded.columns)
```

```python
new_df = new_df.drop(df.select_dtypes(include=['object']),axis=1)
new_df.isna().sum()[new_df.isna().sum() >0]
```

```python
Mode_columns = ['GarageCars','GarageYrBlt','BsmtFullBath','BsmtHalfBath']
Mean_columns = ['LotFrontage', 'MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF',
                'TotalBsmtSF','GarageArea']
for i in Mode_columns:
    new_df[i] =  new_df[i].fillna(new_df[i].mode()[0])
for i in Mean_columns:
    new_df[i] = new_df[i].fillna(np.round(new_df[i].mean()))
new_df.isna().sum()[new_df.isna().sum() >0]
```

```python
training_data = new_df[0:len(df1)]
testing_data = new_df[len(df1):]
testing_data = testing_data.drop(columns='SalePrice')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x = training_data.drop(columns='SalePrice')
y = training_data['SalePrice']
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size = 0.2)
Y_train = np.reshape(Y_train,(-1,1))
Y_test = np.reshape(Y_test,(-1,1))
X_train.shape, Y_train.shape
```

```python
model_1 = LinearRegression()
model_1.fit(x,y)
y_pred = model_1.predict(X_test)
mean_squared_error(Y_test,y_pred)
```

```python
model_2 = XGBRegressor(n_estimators=1000,learning_rate=0.1)
model_2.fit(x,y)
y_pred = model_2.predict(X_test)
mean_squared_error(Y_test,y_pred)
```

```python
model_3 = RandomForestRegressor(n_estimators= 100)
model_3.fit(x,y)
y_pred = model_3.predict(X_test)
mean_squared_error(Y_test,y_pred)
```

```python
sort_indices= np.argsort(Y_test)


plt.figure(figsize=(10,6))
plt.plot(np.arange(len(Y_test)), Y_test, label='Original')  # Note: 'Original' spelling corrected
plt.plot(np.arange(len(Y_test)), y_pred, label='Predicted')  # Corrected here
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
pred = model_2.predict(testing_data)

final = pd.DataFrame()
final['Id'] = testing_data.index
final['SalePrice'] =pred
final.to_csv('output.csv')
```
